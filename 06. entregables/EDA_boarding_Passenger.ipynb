```python
df_1 = runQuery('''
SELECT TO_CHAR(fecharegistro, 'YYYYMMDD') AS date,
       TO_CHAR(fecharegistro, 'Day') AS day,
       TO_CHAR(fecharegistro, 'HH24') AS hour,
       subendelantera + subentrasera AS alightings
FROM source.passenger_route_vehicle
''')
df_1
```

```python
df_2 = runQuery('''
WITH short AS(
        SELECT TO_CHAR(fecharegistro, 'YYYYMMDD') AS date,
               TO_CHAR(fecharegistro, 'Day') AS day,
               TO_CHAR(fecharegistro, 'HH24') AS hour,
               subendelantera + subentrasera AS Subida
        FROM dw.dataset_passanger_route_vehicle
        ), short2 AS(
        SELECT date, day, hour, SUM(Subida) AS Subidas
        FROM short
        GROUP BY date, day, hour
        ORDER BY date
        )
SELECT day, hour, AVG(Subidas) as avg_volumen
FROM short2
GROUP BY day, hour
ORDER BY day, hour
''')
df_2
```

```python
df_2['day'] = df_2['day'].apply(lambda x: x.strip())
df_2['avg_volumen'] = df_2['avg_volumen'].astype('float64')
print(df_2.shape)
```

```python
df_3 = pd.DataFrame(df_2)
```

```python
df_3 = df_2.groupby('day').max()
df_3['avg_volumen'] = df_3['avg_volumen'].astype(float)
df_3["avg_volumen"]
print(df_3.plot.bar(color = "b"))
```

```python
# 'lightsteelblue' 'midnightblue'
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.figure(figsize=(18,22))
for i, day in enumerate(days):
    plt.subplot(4,2,i+1)
    sns.barplot(x="hour", y="avg_volumen", color='#5d535e',
                data=df_2[df_2['day'] == day])
    plt.title(day)
    plt.xlabel("Hour")
    plt.ylabel("Average passengers volume")
    plt.ylim(0, 25000)
    sns.lineplot(x="hour", y="avg_volumen", color='#ff420e', linewidth = 2.5,
                data=df_2[df_2['day'] == day])
```
